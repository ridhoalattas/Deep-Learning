#Function

```
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.legend()
  plt.grid(True)
```

#OOP Import

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.optimizers import Nadam
from tensorflow.keras.layers import BatchNormalization
```

#Multilayer Preceptron (Neural Network)

**Optimizer = SGD**

```
base_model = Sequential()

base_model.add(Flatten(input_shape=(28, 28, 1)))
base_model.add(Dense(128, activation='relu', kernel_initializer='he_uniform')) # Activation relu dipakai, ini salah satu function yang paling banyak dipakai
base_model.add(Dense(10, activation='softmax')) # Activation softmax dipakai untuk classification, nilai 10 karena terdapat 10 class (0-9)

opt = SGD(lr=0.01, momentum=0.9) # Optimizer (Stochastic Gradient Descent), untuk mencari minima dari grafik fungsi loss # lr untuk longkapan data yg akan di check

base_model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy']) # Loss Function

base_model.summary()

history = base_model.fit(feature_train, label_train, batch_size=32, validation_data=(feature_test, label_test), epochs=50)
```

```
history_dataframe = pd.DataFrame(history.history)
history_dataframe['epoch'] = history.epoch
history_dataframe
```

```
plot_loss(history)
```

```
loss, accuracy = base_model.evaluate(feature_test, label_test, verbose=2)

print("Test accuracy : ", accuracy)
print("Test loss : ", loss)
```

#Deeper Convolutional Neural Network (MNIST)

```
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import pandas as pd
```

```
(feature_train, label_train), (feature_test, label_test) = mnist.load_data()

print("Train", feature_train.shape, label_train.shape)
print("Test", feature_test.shape, label_test.shape)

plt.imshow(feature_train[100], cmap=plt.get_cmap('gray'))
```

```
feature_train = feature_train.reshape((feature_train.shape[0], 28, 28, 1))
feature_test = feature_test.reshape((feature_test.shape[0], 28, 28, 1))
```

```
label_train = to_categorical(label_train)
label_test = to_categorical(label_test)
```

```
feature_train = feature_train.astype('float32')
feature_test = feature_test.astype('float32')

feature_train = feature_train / 255.0
feature_test = feature_test / 255.0
```

**Optimizer = SGD**

```
model_deeper_mnist = Sequential()

model_deeper_mnist.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
model_deeper_mnist.add(MaxPooling2D((2, 2)))
model_deeper_mnist.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform'))
model_deeper_mnist.add(MaxPooling2D((2, 2)))

model_deeper_mnist.add(Flatten())

model_deeper_mnist.add(Dense(100, activation='relu', kernel_initializer='he_uniform'))
model_deeper_mnist.add(Dense(10, activation='softmax'))

opt = SGD(learning_rate=0.01, momentum=0.9)

model_deeper_mnist.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

model_deeper_mnist.summary()

history_deeper_mnist = model_deeper_mnist.fit(feature_train, label_train, batch_size=32, validation_data=(feature_test, label_test), epochs=50)
```

```
plot_loss(history_deeper_mnist)
```
